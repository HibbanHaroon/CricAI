```python
from ultralytics import YOLO
```

```python
model = YOLO('best.pt')
```

```python
source = 'videos/video4.mp4'
destination = 'clipped-videos/video4/'
```

```python
results = model(source, stream=True, conf=0.25, classes=[1, 2], save=True, save_frames=True, name=destination)
```
